```python
# notebooks - for experiments, analysis, charts, explanations
import pandas as pd
import numpy as np
#prices contains Close,high,low,open,volume of the 30 stocks

prices = pd.read_csv(
    "../data/processed/stock_prices_clean.csv",
    parse_dates=["Date"]
)

prices.head()
```

```python
prices.info() #2897 trading days (11.5 yrs of data), 30 stocks
```

```python
#Create stock closing prices only 
close_columns = [] #to store close columns 
for col in prices.columns:  #prices contains Close,high, low, open,volume
    if "Close" in col:
        close_columns.append(col) #['Close_AAPL', 'Close_AMZN', 'Close_AVGO', ....]
close_prices = prices[close_columns]
print(close_prices.head())
```

```python
close_prices.shape #(2897, 30), 2897 rows = trading days, 30 columns = stocks 
```

```python
#Calculate Daily returns 
#Convert prices into percentage changes - return = (today price - yesterday price) / yesterday price
apple_price_head = prices["Close_AAPL"].head()
print(apple_price_head)
#Convert prices into percentage changes - day 1 = 100, day 2 = 105, percentage change = 5%
apple_pct_head = apple_price_head.pct_change()
print("pct head change: \n")
print(apple_pct_head)
```

```python
#Calculate Daily returns these are the percentage changes
#Convert prices into percentage changes - return = (today price - yesterday price) / yesterday price
#Convert prices into percentage changes - day 1 = 100, day 2 = 105, percentage change = 5%
close_prices_percentage_change = close_prices.pct_change()
print(close_prices_percentage_change.head())
```

```python
#Remove any rows with NaN values (first row will be NaN because there is no previous day to compare to)
close_prices_percentage_change = close_prices_percentage_change.dropna()
print(close_prices_percentage_change.head())
print("Shape of close_prices_percentage_change: ", close_prices_percentage_change.shape) #(2896, 30), 2896 rows = trading days, 30 columns = stocks
```

```python
#Mometum factor
# Measures how much each stock has increased or decreased over the past 6 months (126 trading days).
#Formula
#Momentum = (Price today / Price 126 trading days ago) - 1
#Higher momentum = stronger recent performance, lower momentum = weaker recent performance
momentum = close_prices.pct_change(periods=126) #126 trading days in 6 months
print(momentum)
momentum.to_csv("../data/processed/momentum.csv")
```

```python
#calculate the volatility 
#measures how much do this stock's daily returns bounce around - standard deviation
volatility = close_prices_percentage_change.std() 
print(volatility)
```
